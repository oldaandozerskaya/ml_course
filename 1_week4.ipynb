##Генерация и визуализация выборки размера COUNT

```
BETA = 0.5
COUNT = 1000
lambda_=1./BETA
x = np.random.exponential(scale=BETA, size=COUNT)
sns.distplot(x)
print('lambda = '+str(lambda_))
```

##Генерация выборок при n = 5, 10, 50

---

Расчёт теоретических параметров распределения.

Вывод графиков для каждого значения n.

Расчёт отклонений от теоретических значений.




```
n = [5, 10, 50]

print('Параметры распределения (на основе теоретических значений):')
mean_theor = 1./lambda_
variance_theor = 1./(lambda_**2)
print('Mean = '+str(mean_theor))
print('Variance = '+str(variance_theor))
print('***')

for ni in n:
  average_values=np.zeros(COUNT)

  for i in range(COUNT):
    x = np.random.exponential(scale=BETA, size=5)
    average_values[i] = np.average(x)

  sns.distplot(average_values, color="deeppink", label="n="+str(ni))
  plt.ylabel('y', fontsize=14)
  plt.xlabel('x', fontsize=14)
  plt.legend()
  plt.show()

  print('Отклонения от теоретических значений, n = ' + str(ni))
  mean = np.average(average_values)
  variance = np.sqrt(np.var(average_values))
  print('Mean = ' + str(mean) + ', отклонение ' + str(math.fabs(mean-mean_theor)))
  print('Variance = ' + str(variance) + ', отклонение ' + str(math.fabs(variance-variance_theor)))
  print('***')
```

##Вывод

---

Графики показывают, что распределение выборочных средних приближается к нормальному. При n=5 (для бета=0.5) наблюдается менее точная аппроксимация по сравнению с n=10 и n=50.
